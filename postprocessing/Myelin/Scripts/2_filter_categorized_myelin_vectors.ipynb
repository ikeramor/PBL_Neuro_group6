```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.colors as mcolors
import matplotlib.patches as patches

import numpy as np
import pandas as pd
import os
```

```python
df_regions = pd.read_csv("/home/jovyan/Desktop/PBL_Neuro/Scripts/myelin_features_all_subjects.csv")
df_networks = pd.read_csv("/home/jovyan/Desktop/PBL_Neuro/Scripts/myelin_features_networks.csv")
df_metadata= pd.read_csv("/home/jovyan/Desktop/PBL_Neuro/eda/OASIS3_metadata_clean.csv")
```

```python
df_regions.Subject_ID.head()
```

```python
df_metadata.Subject_ID.head()
```

```python
#remove sub- so that subject ids match
df_networks["Subject_ID"] = df_networks["Subject_ID"].str.replace("sub-", "", regex=False)
df_regions["Subject_ID"] = df_regions["Subject_ID"].str.replace("sub-", "", regex=False)
df_regions.Subject_ID.head()
```

```python
# Derive 'group' column from 'DEMENTED' and 'NORMCOG'
df_metadata['group'] = np.where(df_metadata['DEMENTED'] == 1, 'DEMENTED', 'NORMCOG')

# Filter myelin vector csv based on the database with the metadata and append 'group' and 'Age' columns
df_networks_filtered = pd.merge(df_networks, df_metadata[['Subject_ID', 'group', 'Age']], on='Subject_ID', how='inner')
df_regions_filtered = pd.merge(df_regions, df_metadata[['Subject_ID', 'group', 'Age']], on='Subject_ID', how='inner')
# Save the filtered file
df_regions_filtered.to_csv('/home/jovyan/Desktop/PBL_Neuro/Scripts/myelin_features_all_subjects_filtered.csv', index=False)
df_networks_filtered.to_csv('/home/jovyan/Desktop/PBL_Neuro/Scripts/myelin_features_networks_filtered.csv', index=False)
```
