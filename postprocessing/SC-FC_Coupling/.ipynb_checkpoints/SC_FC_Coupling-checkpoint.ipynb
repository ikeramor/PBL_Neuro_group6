## Create Network Directory

```python
# Load Schaefer 2018 atlas data (300 ROIs, 7 networks, 2 mm resolution):
atlas = fetch_atlas_schaefer_2018(n_rois=300, yeo_networks=7, resolution_mm=2)

# Define atlas image filepath:
atlas_path = "../../templateflow/tpl-MNI152NLin2009cAsym/tpl-MNI152NLin2009cAsym_res-02_atlas-Schaefer2018_desc-300Parcels7Networks_dseg.nii.gz"

# Load atlas data:
atlas_img = nib.load(atlas_path)
atlas_node_data = atlas_img.get_fdata().astype(int)

# Get the corresponding node labels excluding the background:
labels = atlas["labels"][1:] # 300 labels corresponding to 300 nodes.

# Define a dictionary with network abbreviations as keys and network full names as values:
network_map = {
    "Vis": "Visual Network (VIN)",
    "SomMot": "Somatomotor Network (SMN)",
    "DorsAttn": "Dorsal Attention Network (DAN)",
    "SalVentAttn": "Salience / Ventral Attention Network (SAN/VAN)",
    "Limbic": "Limbic Network (LIN)",
    "Cont": "Control / Frontoparietal Network (FPN/CON)",
    "Default": "Default Mode Network (DMN)"
}

# Initialize an empty dictionary to store results:
network_dict = defaultdict(lambda: {"name": None, "indices": [], "LH": [], "RH": []})

# Iterate through the labels in the atlas:
for roi_idx, label in enumerate(labels, start=1):

    # Get the different text parts of the node label:
    parts = label.split("_")

    # Get the hemisphere of the current node label:
    hemi = parts[1]

    # Get the abbreviation name of the network the current node belongs to:
    network = parts[2]

    # Save both the network name and the index of the label: 
    network_dict[network]["name"] = network_map[network]
    network_dict[network]["indices"].append(roi_idx)

    # Save hemisphere information:
    if hemi == "LH":
        network_dict[network]["LH"].append(roi_idx)
    else:
        network_dict[network]["RH"].append(roi_idx)

# Turn the dictionary into a normal one:
network_dict = dict(network_dict)

# Print summary:
for net, info in network_dict.items():
    print("\n========================")
    print(net)
    print(info["name"])
    print("Total ROIs:", len(info["indices"]))
    print("LH ROIs:", len(info["LH"]))
    print("RH ROIs:", len(info["RH"]))
```

## Compute Myelin-FC Coupling Network Vectors

```python
# Define a dictionary with network abbreviations as keys and network full names as values:
network_map = {
    "Vis": "Visual Network (VIN)",
    "SomMot": "Somatomotor Network (SMN)",
    "DorsAttn": "Dorsal Attention Network (DAN)",
    "SalVentAttn": "Salience / Ventral Attention Network (SAN/VAN)",
    "Limbic": "Limbic Network (LIN)",
    "Cont": "Control / Frontoparietal Network (FPN/CON)",
    "Default": "Default Mode Network (DMN)"
}

# Get a list of network name abbreviations:
networks = list(network_map.keys())

# Load myelin and FC .csv files:
df_myelin = pd.read_csv("../Myelin/Scripts/myelin_features_networks_filtered.csv")
df_fc = pd.read_csv("../FC/fc_metrics.csv")

# Rename column subject ID in fc dataframe:
df_fc = df_fc.rename(columns={"Subject ID": "Subject_ID"})

# Get valid subject IDs from myelin dataframe:
valid_subjects = df_myelin["Subject_ID"].unique()

# Filter FC dataframe
df_fc = df_fc[df_fc["Subject_ID"].isin(valid_subjects)].copy()

# Keep identifier columns:
base = df_fc[['Subject_ID', 'group']].copy()
base = base.reset_index(drop=True)

# Select the wanted columns in the fc dataframe:
wd_cols = [col for col in df_fc.columns if "Weighted_Degree" in col]
fc_part = df_fc[wd_cols]

# Select the wanted columns in the myelin dataframe:
myelin_cols = [col for col in df_myelin.columns if "myelin" in col]
myelin_part = df_myelin[myelin_cols]

# Element-wise multiplication:
multiplied = fc_part.values * myelin_part.values

# Build myelin-fc coupling dataframe with proper column names:
multiplied_df = pd.DataFrame(multiplied, columns=[f"fc_myelin_coupling_{net}" for net in networks])

# Final dataframe
myelin_fc_df = pd.concat([base, multiplied_df], axis=1)

# Save the created Myelin-FC coupling vector as a .csv file:
myelin_fc_df.to_csv("Myelin_FC_coupling_vectors.csv", index=False)

# Compute clinical group average Myelin-FC coupling values per network:
myelin_fc_group_means = (myelin_fc_df.groupby("group")[multiplied_df.columns].mean().round(6))

# Compute difference between groups:
myelin_fc_group_means.loc["DIFFERENCE"] = (myelin_fc_group_means.loc["NORMCOG"] - myelin_fc_group_means.loc["DEMENTED"]).round(6)

# Print results:
print("=== Group Myelin-FC Coupling Averages by Network ===\n")
print(myelin_fc_group_means.to_string())

# Save the created clinical group average Myelin-FC coupling vector as a .csv file:
myelin_fc_group_means.to_csv("Myelin_FC_coupling_vectors_group.csv")
```

## Boxplots of the Network Myelin-FC Coupling for Different Clinical Groups

```python
# Helper function to plot boxplots:
def plot_fc_boxplots(df_fc, df_fc_group, metric, ylabel, networks, network_map):

    # Define network color mapping:
    dark_network_colors = {
        "Vis": "#3700D4",
        "SomMot": "#2462B3",
        "DorsAttn": "#3A6D1C",
        "SalVentAttn": "#6F43B2",
        "Limbic": "#BEF264",
        "Cont": "#DD1952",
        "Default": "#F8A520"
    }

    # Lighter color palette for fills:
    light_network_colors = {
        "Vis": "#9A8CFF",
        "SomMot": "#8BB9E0",
        "DorsAttn": "#8DAA5B",
        "SalVentAttn": "#B39DDB",
        "Limbic": "#DDF7A1",
        "Cont": "#F26A8D",
        "Default": "#FFD27A"
    }

    # Customize the appearance of the outliers:
    flierprops = dict(marker='o', markerfacecolor='#F8C3C6', markeredgecolor='#D32F2F', markeredgewidth=2, markersize=8, linewidth=2)

    # Set plotting style:
    sns.set(style="whitegrid")

    # Define output folder path:
    output_folder = f"Figures/Boxplots/"

    # Make sure output folder exists:
    os.makedirs(output_folder, exist_ok=True)

    # Iterate through networks:
    for net in networks:

        # Extract specific column name:
        col = f"{metric}_{net}" 

        # Create a new figure:
        fig, ax = plt.subplots(figsize=(6,6))

        # Define light fill palette:
        palette = {"CN": light_network_colors[net], "AD": light_network_colors[net]}

        # Create boxplot:
        sns.boxplot(data=df_fc, x="group_plot", y=col, hue='group_plot', palette=palette, width=0.6, linewidth=1.8, flierprops=flierprops, ax=ax)

        # Create background distribution violin plot:
        sns.violinplot(data=df_fc, x="group_plot", y=col, hue='group_plot', palette=palette, inner=None, cut=0, linewidth=1.2, alpha=0.25, ax=ax)

        # Overlay subject points:
        sns.stripplot(data=df_fc, x="group_plot", y=col, color=dark_network_colors[net], jitter=True, size=5, alpha=0.8, marker='o', edgecolor="auto", linewidth=0.7, ax=ax)

        # Extract the metric value of each clinical group for the current network:
        cn_val = df_fc_group[col]['NORMCOG']
        ad_val = df_fc_group[col]['DEMENTED']

        # Define legend elements:
        legend_elements = [Line2D([], [], linestyle='None', markersize=0, label=f'CN Mean: {cn_val:.3f}'), 
                           Line2D([], [], linestyle='None', markersize=0, label=f'AD Mean: {ad_val:.3f}')]

        # Add legend:
        ax.legend(handles=legend_elements, loc="upper right", handlelength=0, handletextpad=0, prop={'weight': 'bold'})

        # Add titles and labels:
        ax.set_title(network_map[net], fontsize=20, fontweight="bold", pad=15)
        ax.set_xlabel("")
        ax.set_ylabel(f"{ylabel}", fontsize=14)

        # Improve spacing:
        plt.tight_layout()

        # Save figure
        fig.savefig(f"{output_folder}/boxplot_FC_{metric}_{net}.svg", dpi=300, bbox_inches="tight")

        # Show plot:
        plt.show()
```

```python
# Create plotting-friendly group labels:
myelin_fc_df["group_plot"] = myelin_fc_df["group"].replace({"NORMCOG": "CN", "DEMENTED": "AD"})

# Plot Myelin-FC boxplots:
plot_fc_boxplots(myelin_fc_df, myelin_fc_group_means, metric="fc_myelin_coupling", ylabel="Myelo-Functional Coupling Across Networks", networks=networks, network_map=network_map)
```

## Compute Group Level Mean Values and Statistics

```python
# Create a helper function to perform Welch's t tests:
def welch_fdr_threshold(df, group_col):
    # Store t-statistics for all columns:
    t_values = []

    # Store p-values for all columns:
    p_values = []

    # Store column names:
    columns = []
    
    # Create an empty list to store the results:
    results = []

    # Select rows belonging to the NORMCOG group:
    normcog = df[df[group_col] == "NORMCOG"]

    # Select rows belonging to the DEMENTED group:
    demented = df[df[group_col] == "DEMENTED"]
    
    # Loop through every column in the dataframe
    for col in df.columns:

        # Skip the grouping column:
        if col == group_col:
            continue

        # Extract values from the NORMCOG group:
        x = normcog[col].dropna()

        # Extract values from the DEMENTED group:
        y = demented[col].dropna()

        # Perform Welch's t-test (unequal variances):
        t_stat, p_value = ttest_ind(x, y, equal_var=False)

        # Save results:
        columns.append(col)
        t_values.append(t_stat)
        p_values.append(p_value)

    # Convert p-values to numpy array
    p_values = np.array(p_values)

    # Apply Benjamini-Hochberg FDR correction
    _, p_fdr, _, _ = multipletests(p_values, alpha=0.05, method="fdr_bh")

    # Create significance mask:
    sig_mask = p_fdr < 0.05

    # Keep t-values only where FDR-significant:
    t_thresholded = np.where(sig_mask, t_values, 0)

    # Return thresholded t-values indexed by column name
    return pd.Series(t_thresholded, index=columns)
```

```python
# Compute FDR corrected and p < 0.05 thresholded Welch's t-Statistic vector for myelo-functional coupling:
myelin_fc_cols = [f"fc_myelin_coupling_{net}" for net in networks]
t_myelin_fc = welch_fdr_threshold(myelin_fc_df[myelin_fc_cols + ["group"]], "group")
```

## Surface Maps Creation

```python
# Import the Schaeffer / Yeo 7 network and 300 parcel atlas:
yeo7_img = nib.load("Atlas/Schaefer2018_300Parcels_7Networks_order.dlabel.nii")
yeo7_data = yeo7_img.get_fdata().squeeze()

# Print the amount of vertices of the parcellation file:
print('Vertices of the parcellation file:', len(yeo7_data))

# Helper function to build Yeo 7 network surface map of the defined metric:
def build_surface_map(labels, networks, network_indices, values):
    # Initialize an empty vector with the size of the atlas:
    surface_map = np.zeros_like(labels, dtype=float)
    
    # Iterate over networks and assign values:
    for i, net in enumerate(networks):
        # Obtain the ROI label values of the current network:
        parcel_ids = network_dict[net]["indices"]

        # Build a mask over the data where values match the current network's label:
        mask = np.isin(labels, parcel_ids)

        # Assign the corresponding NORMCOG value:
        surface_map[mask] = values[i]

    return surface_map
```

```python
# Prepare mean myelin-FC coupling values for both clinical groups of NORMCOG and DEMENTED:
mean_myelin_fc_normcog_values = list(myelin_fc_group_means[myelin_fc_cols].iloc[1])
mean_myelin_fc_demented_values = list(myelin_fc_group_means[myelin_fc_cols].iloc[0])

# Compute maximum and minimum values for plotting:
min_mean_myelin_fc = min(min(mean_myelin_fc_normcog_values), min(mean_myelin_fc_demented_values))
max_mean_myelin_fc = max(max(mean_myelin_fc_normcog_values), max(mean_myelin_fc_demented_values))

# Build mean myelin-FC coupling values and t-statistic maps for both clinical groups of NORMCOG and DEMENTED:
mean_myelin_fc_normcog_map = build_surface_map(yeo7_data, networks, network_dict, mean_myelin_fc_normcog_values)
mean_myelin_fc_demented_map = build_surface_map(yeo7_data, networks, network_dict, mean_myelin_fc_demented_values)
mean_myelin_fc_t_map = build_surface_map(yeo7_data, networks, network_dict, list(t_myelin_fc))
```

## Surface Maps Plotting

```python
# Helper function to create surface brain plots:
def surface_brain_plot(scalar_map, title=None, cmap='YlOrRd_r', cbar_label=None, vmin=None, vmax=None):

    # Proceed if plotting limits are not defined:
    if vmin is None or vmax is None:
        # Compute 2nd and 98th percentiles:
        p2 = np.percentile(scalar_map, 2)
        vmin = p2

        p98 = np.percentile(scalar_map, 98)
        vmax = p98
        
    # Fetch FreeSurfer-like fsLR surface meshes:
    surfaces = fetch_fslr()
    
    # Extract left and right inflated cortical surfaces for plotting geometry:
    lh, rh = surfaces['inflated']

    # Extract sulcal depth maps (used as anatomical background shading):
    sulc_lh, sulc_rh = surfaces['sulc']

    # Initialize a Surfplot object with left and right hemisphere surfaces:
    p = Plot(surf_lh=lh, surf_rh=rh)    

    # Add the main scalar data layer:
    p.add_layer(scalar_map, cmap=cmap, color_range=(vmin, vmax), cbar_label=cbar_label)

    # Build the final figure object from all added layers:
    fig = p.build()

    # Set figure title:
    if title:
        fig.suptitle(title, fontsize=16)

    return fig
```

```python
# Define output figure folder:
output_folder = "Figures/Brain_Surface"

# Make sure output folder exists:
os.makedirs(output_folder, exist_ok=True)

# Create figures without warnings:
with suppress_vtk_warnings():
   
    # Figure 1: Cognitively Normal (CN)
    fig1 = surface_brain_plot(mean_myelin_fc_normcog_map, title=r"Cognitively Normal ($\mathbf{CN}$) Average Myelo-Functional Coupling Across Networks", cmap='coolwarm', cbar_label='AU', vmin=min_mean_myelin_fc, vmax=max_mean_myelin_fc)
    fig1.savefig(f"{output_folder}/Myelo_Functional_Coupling_surface_brain_plot_CN.svg", dpi=300, bbox_inches="tight")
    fig1.show()
    
    # Figure 2: Alzheimer's Disease (AD)
    fig2 = surface_brain_plot(mean_myelin_fc_demented_map, title=r"Alzheimer's Disease ($\mathbf{AD}$) Average Myelo-Functional Coupling Across Networks", cmap='coolwarm', cbar_label='AU', vmin=min_mean_myelin_fc, vmax=max_mean_myelin_fc)
    fig2.savefig(f"{output_folder}/Myelo_Functional_Coupling_surface_brain_plot_AD.svg", dpi=300, bbox_inches="tight")
    fig2.show()

    # Figure 3: Difference
    fig3 = surface_brain_plot(mean_myelin_fc_normcog_map - mean_myelin_fc_demented_map, title=r"$\mathbf{CN}$ - $\mathbf{AD}$ Average Myelo-Functional Coupling Across Networks", cmap='coolwarm', cbar_label='AU', vmin=min_mean_myelin_fc, vmax=max_mean_myelin_fc)
    fig3.savefig(f"{output_folder}/Myelo_Functional_Coupling_surface_brain_plot_Diff.svg", dpi=300, bbox_inches="tight")
    fig3.show
    
    # Figure 4: t-Statistic
    fig4 = surface_brain_plot(mean_myelin_fc_t_map, title=r"$\mathbf{CN}$ vs $\mathbf{AD}$ Average Myelo-Functional Coupling t-Statistic Map (FDR-Corrected)", cmap='RdYlGn', cbar_label='t-Statistic')
    fig4.savefig(f"{output_folder}/Myelo_Functional_Coupling_surface_brain_plot_t.svg", dpi=300, bbox_inches="tight")
    fig4.show
```
