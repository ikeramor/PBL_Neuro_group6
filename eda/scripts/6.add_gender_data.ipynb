# Add Gender Data to the Metadata

```python
# Standard library imports:
from pathlib import Path
import pandas as pd
```

```python
# Load cleaned metadata:
df_meta = pd.read_csv("../OASIS3_metadata_clean.csv")

# Load subject demographics A1 metadata:
df_demo = pd.read_csv("../OASIS3_demographics.csv")

# Keep only the subjects to be processed:
df_demo_filtered = df_demo[df_demo["OASISID"].isin(df_meta["Subject_ID"])]

# Keep only needed columns from df_demo:
df_demo_sub = df_demo[["OASISID", "GENDER", "EDUC"]]

# Merge into df_meta on ID columns:
df_meta = df_meta.merge(df_demo_sub, left_on="Subject_ID", right_on="OASISID", how="left")

# Drop redundant key and rename columns:
df_meta = df_meta.drop(columns=["OASISID"]).rename(columns={"GENDER": "Gender", "EDUC": "Education"})

# Save final metadata:
df_meta.to_csv("../OASIS3_metadata_final.csv", index=False)
```
