# Keep a Single Functional BOLD File per Subject and Session

```python
# Import necessary libraries:
import os
import re
from pathlib import Path

# Define the base derivatives dataset path:
base_path = Path("../../dataset/derivatives")

# List all directories in the base path (subjects) and sort them:
subjects = sorted([d for d in base_path.iterdir() if d.is_dir() and d.name.startswith("sub-")])

# Compile a regex pattern to extract run numbers from filenames:
run_pattern = re.compile(r"run-(\d+)")

# Flag to indicate dry run mode (no actual deletion):
dry_run = True

# Loop over each subject directory:
for subj in subjects:
    # Get all functional session directories for the subject:
    func_path = subj.glob("ses-*/func/")
    
    # Sort session paths to maintain consistent order:
    func_paths = sorted(func_path)
    
    # Initialize variable to store the run number to keep:
    run_to_keep = None

    # Loop over session paths to find the first preprocessed BOLD file:
    for fpath in func_paths:
        # List all preprocessed BOLD files for this subject:
        bold_files = list(fpath.glob(f"{subj.name}_*_desc-preproc_bold.nii.gz"))

        if bold_files:
            # Extract the run number from the first matching file:
            m = run_pattern.search(bold_files[0].name)

            if m:
                # Store the run number to keep:
                run_to_keep = m.group(1)
                
                # Stop searching after finding the first valid run:
                break

    # Skip the subject if no preprocessed BOLD files were found:
    if not run_to_keep:
        print(f"No preprocessed BOLD file found for {subj.name}, skipping...")
        continue

    # Print which run will be kept for this subject:
    print(f"\nSubject {subj.name}, keeping run-{run_to_keep}")
    
    # Loop again through all functional files to delete runs not kept:
    for fpath in func_paths:
        for file in fpath.iterdir():
            # Check if the file has a run number:
            m = run_pattern.search(file.name)
            # If the file belongs to a run not kept, mark it for deletion:
            if m and m.group(1) != run_to_keep:
                if dry_run:
                    # In dry run mode, just print what would be deleted:
                    print(f"[DRY RUN] Would delete: {file}")
                else:
                    # Actually delete the file:
                    file.unlink()
```

```python
# Import necessary libraries:
from pathlib import Path
from collections import defaultdict

# Define the root derivatives folder:
root = Path("../../dataset/derivatives")  

# Initialize a dictionary to store files for each subject:
subject_files = defaultdict(list)  

# Find all relevant files recursively:
for f in root.rglob("*desc-preproc_bold.nii.gz"):
    # Extract the subject ID from the file path:
    subject = next((p for p in f.parts if p.startswith("sub-")), None)

    if subject:
        # If a subject ID was found, append this file path to the list for that subject.
        subject_files[subject].append(f)

# Define a function to sort subjects numerically:
def subject_key(sub):
    return int(sub.split("OAS")[-1])

# Display the results:
for sub in sorted(subject_files, key=subject_key):
    # Sort the subjects using the numeric key defined above.
    print(f"\n{sub}")
    
    # Print the subject ID with a newline for readability.
    for file in sorted(subject_files[sub]):
        
        # Sort the files for each subject alphabetically (or by full path).
        size_gb = file.stat().st_size / (1024**3)
        
        # Print the file path and its size in GB with 2 decimal places.
        print(f"   {file}   |   {size_gb:.2f} GB")  
```
