# Explore Anatomical and Functional File Types in the Dataset

```python
# Import necessary libraries:
import os
from collections import Counter

# Define the root directory of the dataset:
root_dir = "../../dataset/raw"

# Initialize a counter to track anatomical file types:
anat_file_types = Counter()

# Count unique anatomical file types in anat folders:
for root, dirs, files in os.walk(root_dir):
    if os.path.basename(root) == "anat":
        for f in files:
            # Only consider BIDS-like files starting with sub-:
            if f.startswith("sub-"):
                # Extract the name of the file:
                name = f

                # Remove file extensions (.nii.gz or .json):
                if name.endswith(".nii.gz"):
                    name = name[:-7]
                elif name.endswith(".json"):
                    name = name[:-5]
                else:
                    continue

                # Split filename into parts by underscore:
                parts = name.split("_")
                
                # Keep everything after the first two parts (sub-<ID> and ses-<session>):
                suffix = "_".join(parts[2:])

                # Increment count of this file type:
                anat_file_types[suffix] += 1

# Print all unique anatomical file types found:
print("Unique anatomical file types found:\n")
for k, v in sorted(anat_file_types.items()):
    print(f"{k:30s} : {v}")
```
