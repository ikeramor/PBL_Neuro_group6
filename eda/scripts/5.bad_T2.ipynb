```python
# Import necessary libraries:
import pandas as pd
from pathlib import Path
```

```python
# Load the final dataframe:
df_final = pd.read_csv('../OASIS3_metadata_clean.csv')

# Make a list of the subejcts whose T2 volumes are faulty:
subjects_bad_T2 = [
    "OAS30956",
    "OAS30967",
    "OAS30963",
    "OAS30944",
    "OAS30954",
    "OAS30949",
    "OAS30724",
    "OAS30082",
]

# Remove subjects whose T2 volumes are faulty:
df_final = df_final[~df_final["Subject_ID"].isin(subjects_bad_T2)]

# Create the final output folder:
log_dir = Path("logs/5.valid_subjects_t2")
log_dir.mkdir(parents=True, exist_ok=True)

# Kept subjects from the cleaned dataframe:
kept_subjects = sorted(df_final["Subject_ID"].astype(str).unique())

# Write deleted subjects log:
with open(log_dir / "deleted_subjects_T2_log.txt", "w") as f:
    for subj in subjects_bad_T2:
        f.write(f"sub-{subj}\n")

# Write kept log:
with open(log_dir / "kept_subjects_T2_log.txt", "w") as f:
    for subj in kept_subjects:
        f.write(f"sub-{subj}\n")

# Create summary log:
total_processed = len(kept_subjects) + len(subjects_bad_T2)
kept_count = len(kept_subjects)
deleted_count = len(subjects_bad_T2)

summary_text = f"""| SUMMARY |
Total subjects processed: {total_processed}
Subjects with correct T2 kept: {kept_count}
Subjects with faulty T2 deleted: {deleted_count}
"""
# Save summaty log:
with open(log_dir / "summary.txt", "w") as f:
    f.write(summary_text)

# Print summary log:
print(summary_text)
```

```python
# Save final CSV:
df_final.to_csv('../OASIS3_metadata_clean.csv', index=False)
```
